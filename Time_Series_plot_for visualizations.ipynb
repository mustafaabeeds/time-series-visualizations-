## Line plot and Scatter Plot

```python
df.plot(color = 'orange')
plt.ylabel('temp')
plt.xlabel('Date')
plt.title('Line Plot')
plt.show()
```

```python
df.plot(style = '.', color= 'blue')
plt.ylabel('Temp')
plt.title('Line Plot')
plt.show()
```
